```python
########class########

class Person:
    name = "Adam"
    
    def __init__(self):
        self.height = 170
    
    def speak(self):
        print("My name is {0}".format(self.name))
    
    def show(self):
        print("I'm a Person")
        
        
class Girl(Person):
    name = "ABC"
    age = 18
    def __init__(self):
        super().__init__()
    
    def speak(self.sentence):
        print("{0}".format(sentence))
        
    def show(self):
        print("I'm a girl")
        
class XJTLUer():
    def speak(self):
        print("I'm a XJTLUer")
        
class Girl_XJTLUer(XJTLUer, Girl):
    pass

sample = Girl_XJTLUer()
sample.speak()
```

```python
########Exception########
# NameError : try to access a variable that is not declared
# ZeroDivisionError : division is zero
# SyntaxError : wrong grammar
# IndexError : index the out of range 
# KeyError : the key does not exist 
# IOError : wrong input or output ( e.g. there is no such a file to load) 
# ArributeError : try to access an unknown attribution of an object

try:
    f = open("file.txt",'r')
except IOError:
    print("There is an IOError")
except ZeroDivisionError:
    print("There is an ZeroDivisionError")
    
try:
    f = open("file.txt",'w')
    f.write("hello world")
except Exception as e:# "Exception" is the basis of all exceptions
    print(e)
    
# else ：if there is not exception in "try" then execute "else"
while true:
    try:
        f = open("file.txt",'w')
        f.write("hello world")
    except Exception as e:# "Exception" is the basis of all exceptions
        print(e)
        print("Try again")
    else:
        print("else sentence: finished")
        break
        
        
try:
    f = open("file.txt",'w')
    f.write("hello world")
except Exception as e:# "Exception" is the basis of all exceptions
    print(e)
    print("Try again")
finally:
    f.closed()
    print("The f is closed")
    



```
